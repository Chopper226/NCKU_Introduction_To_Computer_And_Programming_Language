迴圈除了用來大量處理項目，另外就是用來檢查用的，所以中間常常會有 if

````python
N = 10
for x in range(N): # 從0開始到N-1
  if x % 2 == 0: #檢查x除以2的餘數是不是0, if 的後面也要有 :
    print(x)    # 也一定要縮排
    print("----")  # 做到同樣縮排的最後一行

  print(f"===={x}=====")
````

1. (2 points) 寫一個包含迴圈的程式，而且中間有if的判斷式，並且加上註解
2. (3 points) 寫一個程式，用for迴圈印出1到N中的質數，而質數的檢查要透過while迴圈來找(小於或等於根號k的數能不能整除k)。不能用到函數或是任何的packages喔。
3. (4 points) 寫一個函數，當$x$是奇數或偶數的時候，給一個式子將$x$換成另一個數字放入$x$。並檢查這個數字有沒有可能變成1。(舉例來說，3n+1問題就是。但這一題是要你寫一個一般化的情況。如果你不知道怎麼寫，可以參考它)。這個式子中會有一些變數，以及初始值$x_0$，要當成函數的輸入。並且測試一下$x_0$在1到N中哪些數字，一直跑下去會變成1。

以下是第一題的答案

````python
x = int( input() ) # 找 1~x 中有幾個奇數幾個偶數

oddNum = 0 # 將奇數個數初始值設為 0
evenNum = 0 # 將偶數個數初始值設為 0

for i in range(1,x+1) :
    if i%2 == 0 : evenNum +=1 # 如果能被 2 整除代表他是偶數 -> 偶數個數+1
    else : oddNum +=1 # 如果不能被 2 整除代表他是奇數 -> 奇數個數+1

print( f"有{oddNum}個奇數" ) # 輸出有幾個奇數
print( f"有{evenNum}個偶數") # 輸出有幾個偶數
````

以下是第二題的答案

````python
n = int( input() )

for i in range(1,n+1) :
    if i == 1 : continue # 1 不是質數也不是合數

    check = 2
    primeNum = True

    while check*check <= i : #check <= sqrt(i)
        if i%check == 0 :
            primeNum = False
            break
        check+=1
    
    if primeNum :
        print( i , end = " ")
````

以下是第三題的答案

````python
def f( x , times ) :
    if x == 1 :
        return times
    if x%3 == 0 :
        return f( x/3 , times+1 )
    elif x%3 == 1:
        return f( 4*x-1 ,times+1 )
    else : return f( 4*x+1 , times+1 ) 


x = int( input() )
for i in range( 1,x+1 ) :
    print( f"{i} : 迭代 {f(i,0)}次後可以得到 1" )
````
